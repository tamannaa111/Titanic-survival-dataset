```python
# 1. Import Libraries
import pandas as pd
from scipy.stats import shapiro, probplot
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Load Data
path = "/content/drive/MyDrive/Ibrahim/tested.csv"
df = pd.read_csv(path)
df.head(5)
```

```python
# 3. Check for Normal Distribution
# For example, let's check the 'Age' column
column = 'Age'
```

```python
# Shapiro-Wilk Test
stat, p_value = shapiro(df[column])
if p_value > 0.05:
    print(f"The {column} column is normally distributed (p-value: {p_value:.4f})")
else:
    print(f"The {column} column is not normally distributed (p-value: {p_value:.4f})")
```

```python
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df[column], kde=True, color='skyblue')
plt.title(f'Histogram of {column}')
plt.xlabel(column)
plt.ylabel('Frequency')
plt.show()
```

```python
# Q-Q Plot
plt.figure(figsize=(8, 6))
probplot(df[column], dist="norm", plot=plt)
plt.title(f'Q-Q Plot of {column}')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Ordered Values')
plt.show()
```
