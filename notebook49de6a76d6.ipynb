```python
train=pd.get_dummies(train,columns=['Sex','Embarked'])
train.head()
```

```python
x=train.drop("Survived",axis=1) ##Feature column
```

# **Data slipting and Selecting best algorithm**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
accuracy_lr=accuracy_score(y_test,y_pred)
print(f"Accuracy of logisticRegresssion is  {accuracy_lr*100:.2f}%")
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
accuracy_rf=accuracy_score(y_test,y_pred)
print(f'Accuracy of RandomForestClassifier is {accuracy_rf*100:.2f}%')
```

# Applying best algorithm and create submission file
